## 2.) Data Modeling Steps <br>
 
 The planning here is to create two dataset. The first dataset is the raw dataset which contain each reaction refering to what type and category it is and also how the scoring in terms of numbers. In conclusion the steps as follows :
 <ol>
    <li> Join React with Content with React as base
    <li> Join React with React_type with React as base
    <li> Export the raw dataset
    <li> do some EDA
  <ol>

```python
react_type.sort_values(by='Score')
```

```python
react.rename(columns={'Content ID':'ContentID'},inplace = True)
content.rename(columns={'Content ID':'ContentID'},inplace = True)
```

```python
query = """
SELECT 
    r.ContentID,
    r."User ID",
    r.Type AS 'Reaction Type',
    r.Datetime,
    rt.Sentiment,
    CASE rt.Sentiment
        WHEN 'negative' THEN -1
        WHEN 'neutral' THEN 0
        ELSE 1 END AS PromoPoint
    ,
    rt.Score,
    c.Type AS 'Content Type',
    c.Category 
 FROM react AS r
 LEFT JOIN react_type rt ON rt.Type = r.Type 
 LEFT JOIN content c ON c.ContentID = r.ContentID;
"""
df = sqldf(query)
```

```python
df
```

## 3) Data Exploration Steps <br>

The exploration is used to answer some business Question. To iterate the case which discussed here is about Social Buzz wants to do an IPO. The case needs to explore how to dig from the big data to optimize the optimum category and genre for social buzz.

```python
#Daily Activity of Platform's Users
df.groupby(pd.to_datetime(react['Datetime']).dt.strftime('%Y,%m')).aggregate({ 'User ID':lambda x: x.nunique()}).plot(
    kind = 'bar',
    figsize = (10,5),
    xlabel = 'Year, Month',
    ylabel = 'Unique Users',
    title = 'Daily Unique Users',
    legend = False
    )
plt.tick_params(rotation=45)
```

```python
df.groupby([pd.to_datetime(react['Datetime']).dt.strftime('%Y,%m'),'Content Type']).ContentID.aggregate('count').unstack().apply(lambda x: x*100/sum(x), axis=1).plot(
    kind="bar", 
    stacked=True,
    figsize = (10,6),
    title = "Contents Compositions in Social Buzz",
    xlabel = "Year, Month",
    ylabel = "Content Compositions (%)",
    xlim = (0,100))
plt.tick_params(rotation=45)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

From the visualization above we can see **That both Content Compostions** and relatively stable. So we can assume there is **No Seasonability** of contents type towards active users and Social Buzz has stable active user. For furhter analysis we should focus further on What those people like based on the **Sentiments** they gives.

```python
#Plotting Compositions with 2 sub plots one to show the distributions with bar chart on each reactions and the
# Tabulate NPV Values
viz4 = df.groupby('Category').aggregate({'PromoPoint':'sum','Score' : 'mean','ContentID':'count'})
viz4['Score'] = viz4['Score'].map('{:,.2f}'.format)
```

```python
viz4.sort_values(['PromoPoint','Score','ContentID'],ascending = False)
```
