## 2.) Data Modeling Steps <br>
 
 The planning here is to create two dataset. The first dataset is the raw dataset which contain each reaction refering to what type and category it is and also how the scoring in terms of numbers. In conclusion the steps as follows :
 <ol>
    <li> Join React with Content with React as base
    <li> Join React with React_type with React as base
    <li> Export the raw dataset
    <li> do some EDA
  <ol>

```python
react_type
```

```python
!pip install pandasql
from pandasql import sqldf
```

```python
react.rename(columns={'Content ID':'ContentID'},inplace = True)
content.rename(columns={'Content ID':'ContentID'},inplace = True)
```

```python
query = """
SELECT 
    r.ContentID,
    r.Type AS 'Reaction Type',
    r.Datetime,
    rt.Sentiment,
    rt.Score,
    c.Type AS 'Content Type',
    c.Category 
 FROM react AS r
 LEFT JOIN react_type rt ON rt.Type = r.Type 
 LEFT JOIN content c ON c.ContentID = r.ContentID;
"""
df = sqldf(query)
```

```python
df.head()
```

```python
query = """
SELECT 
    df.Category,
    SUM(df.Score) as 'Total Score'
 FROM df
 GROUP BY 1 ORDER BY 2 DESC Limit 5;
"""
df2 = sqldf(query)
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    df.to_excel(writer, sheet_name='Clean Data', index=False)
    df2.to_excel(writer, sheet_name='Top 5 Category', index=False)
```
